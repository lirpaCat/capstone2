<b>Fixing PDF Files for Tables</b>

```python
import csv

import numpy as np
import networkx as nx 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Select some data to work with for this session

#first create list using all columns
allData = []
with open('18000.csv', 'r') as f:
    filereader = csv.reader(f, delimiter=",", quotechar='"')
    #next(filereader) # skips header row, we don't want to skip since we don't have a header
    for row in filereader:
            allData.append(row)
 
#fix the solver column by separating out responders, whitespace, and formatting unknowns
fixedSolver = []
edge = []

unknown = {"", "Unknown", "unknown"}
for x in allData:
    if "\n" in x[4]:
        solvers = x[4].split("\n")
        for y in solvers:
            one=x[3].replace(' ', '')
            two=y.replace(' ', '')
            if(one in unknown):
                one="UNKNOWN"
            if(two in unknown):
                two="UNKNOWN"
            fixedSolver.append([x[0], x[1], x[2], one,two, x[5], x[6], x[7], x[8]])
    else:
        one=x[3].replace(' ', '')
        two=x[4].replace(' ', '')
        if(one in unknown):
            one="UNKNOWN"
        if(two in unknown):
            two="UNKNOWN"
        fixedSolver.append([x[0], x[1], x[2], one,two, x[5], x[6], x[7], x[8]])
        
fixedPDF = []

unknown = {"", "Unknown", "unknown"}
for x in fixedSolver:
    if "\n" in x[3]:
        authors = x[3].split("\n")
        for y in authors:
            one=y.replace(' ', '')
            two=x[4].replace(' ', '')
            if(one in unknown):
                one="UNKNOWN"
            if(two in unknown):
                two="UNKNOWN"
            fixedPDF.append([x[0], x[1], x[2], one,two, x[5], x[6], x[7], x[8]])
    else:
        one=x[3].replace(' ', '')
        two=x[4].replace(' ', '')
        if(one in unknown):
            one="UNKNOWN"
        if(two in unknown):
            two="UNKNOWN"
        fixedPDF.append([x[0], x[1], x[2], one,two, x[5], x[6], x[7], x[8]])
```

```python
df = pd.DataFrame(fixedPDF)
df.to_csv('18000a.csv')
```

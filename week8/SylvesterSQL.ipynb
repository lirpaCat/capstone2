```python
#create edges from the graph
edges = []
with open('sylvester.csv', 'r') as f:
    filereader = csv.reader(f, delimiter=",", quotechar='"')
    next(filereader) # skips header row
    for row in filereader:
            edges.append(row)
```

```python
edges
```

<b>That's a lot of rows.  Try a new graph with the types!</b>

```python
types = []
for x in edges:
    types.append(x[2])
```

```python
types
```

<b>Make a histogram of all the types, then adjust the size and save as a png in our current directory!</b>

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(30, 2))
plt.hist(types, bins=100)
plt.savefig('histogram.png', format="PNG")
```
